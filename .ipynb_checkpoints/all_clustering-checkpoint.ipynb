```r
df_white <- df_all[df_all$type == 'w', ]
df_white$type <- NULL
```

```r
summary(df_white)
```

```r
library(factoextra)
library(FactoMineR)
library(tidyverse)
library(NbClust)
library(ggpubr)
```

```r
res_white.pca <- PCA(df_white[, 1:11], scale.unit = TRUE, ncp = 7)
```

```r
scores <- get_pca_ind(res_white.pca)$coord
loadings <- get_pca_var(res_white.pca)$coord
```

```r
g1 <- ggplot(as.data.frame(scores), aes(x = Dim.1, y = Dim.2)) +
        geom_point(aes(fill = df_white$quality), color = 'black', shape = 21, size = 3) +
        geom_hline(yintercept = 0, linetype = "solid", color = "black", size = .2) +
        geom_vline(xintercept = 0, linetype = "solid", color = "black", size = .2) +
        theme_bw() + 
        theme(legend.position = 'top') +
        labs(title = 'PC1 vs PC2', fill = 'Quality') +
        xlab('PC1') + 
        ylab('PC2')
g2 <- ggplot(as.data.frame(scores), aes(x = Dim.1, y = Dim.3)) +
        geom_point(aes(fill = df_white$quality), color = 'black', shape = 21, size = 3) +
        geom_hline(yintercept = 0, linetype = "solid", color = "black", size = .2) +
        geom_vline(xintercept = 0, linetype = "solid", color = "black", size = .2) +
        theme_bw() + 
        theme(legend.position = 'none') +
        labs(title = 'PC1 vs PC3') +
        xlab('PC1') + 
        ylab('PC3')
g3 <- ggplot(as.data.frame(scores), aes(x = Dim.2, y = Dim.3)) +
        geom_point(aes(fill = df_white$quality), color = 'black', shape = 21, size = 3) +
        geom_hline(yintercept = 0, linetype = "solid", color = "black", size = .2) +
        geom_vline(xintercept = 0, linetype = "solid", color = "black", size = .2) +
        theme_bw() + 
        theme(legend.position = 'none') +
        labs(title = 'PC2 vs PC3') +
        xlab('PC2') + 
        ylab('PC3')
g4 <- ggplot(as.data.frame(scores), aes(x = Dim.1, y = Dim.4)) +
        geom_point(aes(fill = df_white$quality), color = 'black', shape = 21, size = 3) +
        geom_hline(yintercept = 0, linetype = "solid", color = "black", size = .2) +
        geom_vline(xintercept = 0, linetype = "solid", color = "black", size = .2) +
        theme_bw() + 
        theme(legend.position = 'none') +
        labs(title = 'PC1 vs PC4') +
        xlab('PC1') + 
        ylab('PC4')

ggarrange(g1, g2, g3, g4, 
            nrow = 2, ncol = 2, align = 'hv', 
            common.legend  = TRUE, legend = 'top')
```
